```python
import numpy as np
import matplotlib.pyplot as plt
```

### 三段式传播模型

```python

def geometrical(distlist):

    r1 = 0.0
    p1 = -0.2
    r2 = 70.0
    p2 = 0.5
    r3 = 150.0
    rr1 = 1.0
    rr2 = 501.0
    nr = 500

    zr = np.zeros(len(distlist), dtype='float')

    counter = 0
    for i in distlist:
        if r1 <= i < r2:
            zr[counter] = 1.0 / i
        elif r2 <= i < r3:
            zr[counter] = 1.0 / r2*((r2/i)**p1)
        elif i >= r3:
            zr[counter] = 1.0 / r2*((r2/r3)**p1)*((r3/i)**p2)
        counter += 1
    return zr  
```

### Q

```python
w = np.arange(0.01, 50, dtype=float)
eta = 0.66;
Qfactor = 186.0
Q  = [Qfactor* f ** eta for f in w]

plt.figure(figsize=(8,6))
plt.plot(w, Q, 'b-',c='purple',lw='3', ls='-')
plt.grid()
plt.show()
```

### site amplification

```python


f1=0.01
f2=100
n=20
a1 = (np.log(f2)-np.log(f1))/(n-1)
w1 = np.arange(0, n, dtype=float)
w1[0] = f1
for i in range(1, int(n)):
    w1[i] = np.exp(np.log(w1[0])+(i*a1))

frequency = np.array([0.01,0.09,0.16,0.51,0.84,1.25,2.26,3.17,6.05,16.6,61.2], dtype=float)
amplification = np.array([1.00,1.10,1.18,1.42,1.58,1.74,2.06,2.25,2.58,3.13,4.00], dtype=float)

z = np.polyfit(np.log(frequency), np.log(amplification), 1)
a = np.poly1d(z)
rocksiteamplification = [np.exp(a(x)) for x in np.log(w1)]

```

```python
plt.figure(figsize=(8,6))
plt.plot(w1, rocksiteamplification, color='#994c00',lw='3', ls='-')
plt.plot(frequency, amplification,'ro')
plt.grid()
plt.show()
```
